## Indexing and Selecting Data in Pandas
Let's separate this dataframe into three new dataframes - one for each metric (mean, standard error, and maximum). To get the data for each dataframe, we need to select the `id` and `diagnosis` columns, as well as the ten columns for that metric.

```python
#view the index number and label for each column
for index, column in enumerate(df.columns):
    print(index, column)
```

### we can select data using loc and iloc, loc uses rows and column to select data, while iloc uses index number

```python
#select all the columns from id to the last mean column
df_means = df.loc[:,'id':'fractal_dimension_mean']
```

This uses the loc method to select all rows and the columns 'id' and 'fractional_dimentional_mean' from the DataFrame df.

```python
df_means.head()
```

## Save the dataframe of mean

```python
df_means.to_csv('Breast_cancer_means.csv',index=False)
```

### Selecting Multiple Ranges in Pandas
Selecting the columns for the mean dataframe was pretty straightforward - the columns we needed to select were all together (id, diagnosis, and the mean columns). Now we run into a little issue when we try to do the same for the standard errors or maximum values. id and diagnosis are separated from the rest of the columns we need! We can't specify all of these in one range.n.

```python
#create a separate dataframe of the column containing standard deviation 
df_sd = df.iloc[:, np.r_[:2,12:22]]
```

[:, np.r_[:2, 12:22]]: This part specifies the rows and columns to be selected. The colon : before the comma represents all rows. The part after the comma specifies the columns. The :2 selects the columns from index 0 to 1 (inclusive), and 12:22 selects columns from index 12 to 21 (inclusive).

```python
#Display the first 5 rows
df_sd.head()
```

## Save the dataframe of standard_deviation

```python
df_sd.to_csv('Breast_cancer_sd.csv',index = False)
```

```python
#create a separate dataframe of the column containing max cols 
df_max = df.iloc[:, np.r_[:2,22:32]]
```
